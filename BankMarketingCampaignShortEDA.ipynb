# EDA

- **During economic crisis :**
    - companies including banks need to spend their capital wisely in order to survive the cycle
    - banks often experience liquidity problems, due to massive numbers of customers cashing-out their money
    - the problems often gets severe, since there might be a spike in non-performing-loan credits
    - therefore, banks need as much deposits as possible from customers, but should keep maintaining it's cost

# Deposits Marketing

- **Offering term deposits to all clients will be inneficient and costly**


- **Clients with certain features, are less potential clients to place a term deposits**


- **Banks need to classify wether the customer has a high or low potential on placing a term deposits**


# Conclusion

**Based on this data:**

- **11.27% of contacted clients are willing to put a term deposits**


- **Elderly group are the most willing to put a term deposits, followed by young adult, senior, and middle age**
    
    Success Rate :
    - 46.84% of elderly group (age 65+)
    - 15.22% of young adult group (age 17-30)
    - 12.07% of senior group (age 51-65)
    - 9.14% of middle age group (age 31-50)


- **Students are the most willing to put a term deposits, followed by retired people, unemployed, and others**
    
    Success Rate :
    - 31.43% of students
    - 25.23% of retired
    - 14.20% of unemployed
    - 12.97% of admin
    - 11.22% of management
    - 10.83% of technician
    - 10.49% of self employed
    - 10% of housemaid
    - 8.52% of entrepreneur
    - 8.14% of services
    - 6.89% of blue collar


- **Person with higher education is more likely to put a term deposits**
    
    Success Rate :
    - 13.00% university degree or equal
    - 10.15% basic 9 year education
    - 9.52%  lower than basic 9 year education


- **Married person is less likely to put a term deposits**
    
    Success Rate :
    - 10.32% divorced (including widowed)
    - 14.00% single
    - 15.00% unstated/unknown


- **Person with no credit default are more likely to put a term deposits**
    
    Success Rate :
    - 12.88% with no credit default
    - 5.15% unknown/unstated
    - 0.00% with credit default


- **Person with housing loan is slightly more likely to put a term deposits**
    
    Success Rate :
    - 11.61% with housing loan
    - 10.88% without housing loan
    - 10.80% unknown/unstated


- **Person who has no loan is slightly more likely to put a term deposits**
    
    Success Rate :
    - 11.34% no loan
    - 10.93% has loan
    - 10.80% unknown/unstated


- **Person who put a term deposits in previous campaign are very likely to put another term deposits** 
    
    Success Rate :
    - 65.11% accepted the previous offer
    - 14.22% rejected the previous offer
    - 8.83% no previous offer


- **Person contacted once is more likely to put a term deposits**
    
    Success Rate :
    - 13.03% contacted once
    - 11.45% contacted twice
    - 10.75% contacted three times
    - 7.27% contacted more than three times


- **Person contacted on other campaign are more likely to put a term deposits**
    
    Success Rate :
    - 58.70% contacted more than twice on other campaign
    - 46.41% contacted twice on other campaign
    - 21.20% contacted once on other campaign
    - 8.83% has never been contacted on other campaign


- **Person who are contacted via cellular are more likely to put a term deposits**
    
    Success Rate :
    - 14.74% via cellular
    - 5.23% via telephone


- **Thursday is the best day to contact customers, followed by Tuesday, Wednesday, Friday and Monday**
    
    Success Rate :
    - 12.12% Thursday
    - 11.78% Tuesday
    - 11.67% Wednesday
    - 10.80% Friday
    - 9.94% Monday


- **1st Quarter of the year, is the best time to contact customers, followed by 4th, 3rd and 2nd **This due to saving rate spike during March, September, October and December**
    
    Success Rate :
    - 50.55% 1st Quarter
    - 16.40% 4th Quarter
    - 11.20% 3rd Quarter
    - 9.13% 2nd Quarter


- **March, September, October, and December are among the best months to contact customers**
- **May, June, and July are among the worst months to contact customers**
    
    Success Rate :

    - 48.90% December

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import xlrd
warnings.filterwarnings('ignore')

%matplotlib inline
```

```python
df = pd.read_csv('BankMarketingEDA.csv')
dfbin = pd.read_csv('BankMarketingEDABin.csv')
```

**Rename for Better Visualization**

```python
df = df.rename(columns={'age': 'Age',
                        'job': 'Job',
                        'marital': 'Marital',
                        'education': 'EducationLevel',
                        'default': 'HasDefault',
                        'housing': 'HousingLoan',
                        'loan': 'HasLoan',
                        'contact': 'ContactType',
                        'month': 'ContactMonth',
                        'day_of_week': 'ContactDay',
                        'campaign': 'ThisCampaignContact',
                        'previous': 'PreviousCampaignContact',
                        'poutcome': 'PreviousOutcome',
                        'euribor3m': 'EuriborRate3Month',
                        'nr.employed': 'NumberOfEmployees',
                        'y': 'Success'
                       })

dfbin = dfbin.rename(columns={'age_bin': 'AgeGroup',
                        'job': 'Job',
                        'alljob': 'AllJob',
                        'marital': 'Marital',
                        'edu_bin': 'EducationLevel',
                        'edu_bin_num': 'EducationLevelNum',
                        'default': 'HasDefault',
                        'housing': 'HousingLoan',
                        'loan': 'HasLoan',
                        'contact': 'ContactType',
                        'month': 'ContactQuarter',
                        'day_of_week': 'ContactDay',
                        'campaign': 'ThisCampaignContact',
                        'campaign_bin': 'ThisCampaignContactBin',
                        'previous': 'PreviousCampaignContact',
                        'previous_bin': 'PreviousCampaignContactBin',
                        'poutcome': 'PreviousOutcome',
                        'euribor3m': 'EuriborRate3Month',
                        'nr.employed': 'NumberOfEmployees',
                        'employed_bin': 'NumberOfEmployeesBin',
                        'y': 'Success'
                       })
```

```python
df['HousingLoan'] = df['HousingLoan'].map({
    0 : 'No',
    1 : 'Unknown',
    2 : 'Yes'
})

df['HasDefault'] = df['HasDefault'].map({
    0 : 'No',
    1 : 'Unknown',
    2 : 'Yes'
})

df['HasLoan'] = df['HasLoan'].map({
    0 : 'No',
    1 : 'Unknown',
    2 : 'Yes'
})

df['PreviousOutcome'] = df['PreviousOutcome'].map({
    0 : 'No',
    1 : 'Not-Exist',
    2 : 'Yes'
})

df['ContactType'] = df['ContactType'].map({
    0 : 'Telephone',
    1 : 'Celullar'
})



dfbin['HousingLoan'] = dfbin['HousingLoan'].map({
    0 : 'No',
    1 : 'Unknown',
    2 : 'Yes'
})

dfbin['HasDefault'] = dfbin['HasDefault'].map({
    0 : 'No',
    1 : 'Unknown',
    2 : 'Yes'
})

dfbin['HasLoan'] = dfbin['HasLoan'].map({
    0 : 'No',
    1 : 'Unknown',
    2 : 'Yes'
})

dfbin['PreviousOutcome'] = dfbin['PreviousOutcome'].map({
    0 : 'No',
    1 : 'Not-Exist',
    2 : 'Yes'
})

dfbin['ContactType'] = dfbin['ContactType'].map({
    0 : 'Telephone',
    1 : 'Celullar'
})
```
